* Importing all the required modules

```python
#importing all the required libraries
import pandas as pd
import numpy as np
import re
import urllib
```

* Reading file from github repository in python environment using urllib library 

```python
# reading file from github
url = "https://raw.githubusercontent.com/surbhinayak10/DataAnalysisProjects/main/Analyzing%20Chess%20Tournament%20Results/tournamentinfo.txt"
resource = urllib.request.urlopen(url).readlines()
resource
```

* Now cleaning the file and decoding in utf-8 format

```python
# decoding data in utf-8 format 
list_clean = []
for i in resource:
    list_clean.append(i.strip().decode("utf-8")) 
list_clean
```

* Removing lines which does not contains any relevant content. Here we can observe that every 3rd line does not have any content so removing it using for loop. 

```python
content_list = []
# removing content if index is divisible by 3 
for i in range(1,len(list_clean)):
    if i%3!=0:
        content_list.append(list_clean[i])
        
# displaying content list
content_list
```

* Now we can observe that every players detail is in 2 lines so here we are extracting content name, number and opposite players index.
* For extracting name, number and opposite player index creating empty list and written regular expression so we can extract information.
* Using for loop for all odd indexes extracting information using findall() function in the list.
* Last storing all extracted data in list.

```python
# regex for extracting players name, number and opposite player index
name = re.compile("[a-zA-Z]+\s[a-zA-Z]+\s[a-zA-Z]*")
number = re.compile("[0-9]\.[0-9]")
opposite = re.compile("[0-9]{1,2}")

# Empty list for storing data
names_list = []
No_of_point_list = []
opp_player = []

#for loop for extracting text for even indexs in list 
for i in range(0,len(content_list),2):
    names = name.findall(str(content_list[i]))
    names_list.append(names[0].strip()) # Cleaning names list if contains any space 
    
    numbers = number.findall(str(content_list[i]))
    if numbers != []:
        No_of_point_list.append(numbers[0])
    
    opposite_player = opposite.findall(str(content_list[i]))
    opp_player.append(opposite_player[3:])
    
# Storing data in desired form 
names_list[0] = "Player's Name"
No_of_point = ['Total Number of Points'] + No_of_point_list
opp_player[0] = 'Opposite players list'

# Displaying all the list content
print(names_list)
print('-------------------------------------------------------------------------------------------------------------')
print(No_of_point)
print('-------------------------------------------------------------------------------------------------------------')
print(opp_player)
```

* Now we can observe that every players detail is in 2 lines so here we are extracting content state, pre-rating and post-rating of players.
* For extracting state, pre-rating and post-rating of players creating empty list and written regular expression so we can extract information.
* Using for loop for all odd indexes extracting information using findall() function in the list and subtracting the unwanted text using sub() function .
* Last storing all extracted data in list.

```python
# Empty list for storing text
state_list = []
players_pre_rat_list = []
players_post_rat_list = []

# Regular expresions for reuired text
state = re.compile("[A-Z]{2}")
players_pre_rating = re.compile("R:\s+[0-9]{1,4}")
players_post_rating = re.compile("\>[0-9]{3,4}")

# collecting data for odd indexs
for i in range(1,len(content_list),2):
    states = state.findall(content_list[i])
    state_list.append(states[0])
    
    players_p_rat = players_pre_rating.findall(content_list[i])
    if players_p_rat != []:
        players_pre_rat_list.append(re.sub('R: ','',players_p_rat[0]))
        
    players_po_rat = players_post_rating.findall(content_list[i])
    if players_po_rat == []:
        players_post_rat_list.append(np.NaN)
    else:
        players_post_rat_list.append(re.sub('>','',players_po_rat[0]))
    
# Storing data in desired form 
state_list[0] = "Player's State"
players_pre_rating = ["Player's Pre-Rating"] + players_pre_rat_list
players_post_rat_list[0] = "Player's Post-Rating"

# Displaying all the list content
print(state_list)
print('-------------------------------------------------------------------------------------------------------------')
print(players_pre_rating)
print('-------------------------------------------------------------------------------------------------------------')
print(players_post_rat_list)
```

* Creating list for player_id so we can use it later as column in dataframe using for loop in list comprehension  

```python
# Creating list for player_id
Player_id = [i for i in range(0,len(names_list))]
Player_id[0] = 'Player_id'
print(Player_id)
```

* Creating dataframe with all the list in desired sequence using DataFrame() function 

```python
df_chess_score = pd.DataFrame({0:Player_id,
                               1:names_list,
                               2:state_list,
                               3:No_of_point, 
                               4:players_pre_rating,
                               5:players_post_rat_list,
                              6:opp_player})

#Displaying dataframe first 5 rows
df_chess_score.head()
```

* Now making zeroth row as column names using iloc() function and dataframe slicing 

```python
# Making zeroth row as header
df_chess_score.columns = df_chess_score.iloc[0]
df_chess_score = df_chess_score[1:]

#Displaying dataframe first 5 rows
df_chess_score.head()
```

```python
# Displaying more information about data set
df_chess_score.info()
```

* Now we are calculating average of opposite players pre-ratings.
* For that first we are collecting list of each players pre-rating using for loop with logic to collect opposite players pre-rating.
* At last creating a list calculating average and creating a list of average pre-rating of each opposite player.

```python
#Creating empty list
Avg_pre_tournmt = []

#For loop for creating list for pre-rating 
for ind, i in enumerate(df_chess_score['Opposite players list'].to_list()):
    lst = []
    for j in i:
        if int(j) in df_chess_score['Player_id'].astype(int).to_list():
            lst.append(int(df_chess_score.loc[int(j),"Player's Pre-Rating"]))
    Avg_pre_tournmt.append(lst)

# Displaying all the list content
print(Avg_pre_tournmt)
print('-------------------------------------------------------------------------------------------------------------')
   
Avg_value = [int(sum(i)/len(i)) for i in Avg_pre_tournmt]
print(Avg_value)
```

* Now adding a list in dataframe with column name 'Average Pre Tournament Chess Rating of Opponents'

```python
df_chess_score = df_chess_score.copy()
df_chess_score["Average Pre Tournament Chess Rating of Opponents"] = Avg_value
# Displaying first 5 rows of dataframe
df_chess_score.head()
```

* Creating final dataframe

```python
# Creating a dataframe with desired columns 
df_chess_score_final = df_chess_score[["Player's Name","Player's State","Total Number of Points", "Player's Pre-Rating",
                                       "Average Pre Tournament Chess Rating of Opponents"]]
# Displaying dataframe
df_chess_score_final
```

* Exporting data in CSV format

```python
# generating a CSV file with structured data 
df_chess_score_final.to_csv('tournamentinfo.csv', index = None)
```
