## 4. Inference

**1. Which Nation has the highest earning over years?**

For answering this question first we are grouping the dataset on 'Nationality' column using groupby() function and taking the total earning from 'Earning_mil' column using sum() function. Then sorting the value in ascending order.

```python
df_ath.groupby('Nationality')['Earning_mil'].sum().sort_values(ascending = False)
```

* Plotting total athletes earning for nations using matplotlib 

```python
ax = plt.axes()
ax.set_facecolor("gainsboro")
df_ath.groupby('Nationality')['Earning_mil'].sum().sort_values(ascending = False).plot(kind = "bar",fontsize=13,
        width=0.8, color = 'navy', legend = False, title = "Total athletes earning for nations", figsize=(14,8),alpha=0.4)
# x and y axis labelling
plt.xlabel('Nations',fontsize=13)
plt.ylabel('Earning in US Dollars(million)',fontsize=13)
plt.xticks(rotation = 60) 
plt.show()#for displaying plot
```

* Plotting total athletes earning for nations using Seaborn

```python
sns.set_theme()
fig_dims = (15,8)
fig, ax = plt.subplots(figsize=fig_dims)
df_total_earning = df_ath.groupby('Nationality')['Earning_mil'].sum().sort_values(ascending=False).reset_index()
plot = sns.barplot(x="Nationality", y="Earning_mil", data=df_total_earning, color = 'navy',ci=None,alpha=0.5)
ax.tick_params(axis='x', rotation=60)
# x and y axis labelling
plt.xlabel('Nations',fontsize=13)
plt.ylabel('Earning in US Dollars(million)',fontsize=13)
plt.title('Total athletes earning for nations',fontsize=13)
plt.show()#for displaying plot
```

From the bar plot we can observe that 'USA' has the highest earning from year 1990 to 2020.

**2. What is the earning trend of highest paid top 5 athletes over the years?**

Here we are taking data of Top 5 highest paid athletes using groupby() and sum() function on 'Name' and 'Earning_mil' column and creating a new dataframe. Then creating a line chart for each athlete.

* Plotting earning trend of highest paid top 5 athletes over the years using matplotlib

```python
fig_dims = (15,8)
fig, ax = plt.subplots(figsize=fig_dims)
ax = plt.axes()
ax.set_facecolor("gainsboro")
df_name = df_ath.groupby(['Name'])['Earning_mil'].sum().sort_values(ascending = True).reset_index().tail(5)

index_lst =[] # empty list for storing indexes
for i in range(df_ath.shape[0]):
    if df_ath['Name'][i] in df_name['Name'].to_list():
        index_lst.append(i)
        
df_new = df_ath.loc[index_lst] #creating a new dataframe with selected indexs
       
for name in df_name.Name.to_list():
    ax.plot(df_new[df_new.Name==name].Year,df_new[df_new.Name==name].Earning_mil,label=name,marker='o')

# x and y axis labelling
ax.set_xlabel("year",fontsize=13)
ax.set_ylabel("Earning_mil",fontsize=13)
ax.legend(loc='best',fontsize=13)
plt.show();#for displaying plot
```

* Plotting earning trend of highest paid top 5 athletes over the years using Seaborn

```python
sns.set_theme()
fig_dims = (15,8)
fig, ax = plt.subplots(figsize=fig_dims)
df_name = df_ath.groupby(['Name'])['Earning_mil'].sum().sort_values(ascending = True).reset_index().tail(5)

index_lst =[] # empty list for storing indexes
for i in range(df_ath.shape[0]):
    if df_ath['Name'][i] in df_name['Name'].to_list():
        index_lst.append(i)

df_new = df_ath.loc[index_lst]  #creating a new dataframe with selected indexs
sns.lineplot(x=df_new['Year'], y=df_new['Earning_mil'] , hue='Name', data=df_new, ax=ax,marker='o')
# x and y axis labelling
ax.set_xlabel("year",fontsize=13)
ax.set_ylabel("Earning_mil",fontsize=13)
ax.legend(loc='best',fontsize=13)
plt.show();#for displaying plot
```

From above line chart, we can notice that multiple line charts represents the earning trend of highest paid top 5 athletes from 1990 to 2020. Here, we can observe that 'Tiger Woods' has been pretty consistently making in the list of highest paid athlete for almost all the years in this dataset.

**3. Which sport produces highest earnings?**

For answering this research question first we are grouping 'Sport' column using groupby() and taking the sum of earning using sum() function from each nation, Then ploting the bar chart using matplotlib and seaborn module.

* Plotting each sport earning using matplotlib

```python
ax = plt.axes()
ax.set_facecolor("gainsboro")
df_ath.groupby('Sport')['Earning_mil'].sum().sort_values(ascending = False).plot(kind = "bar", width=0.8,fontsize=13,
            color = 'darkorchid', legend = False, title = "Total produced earning for each sport", figsize=(14,8),alpha=0.5)
# x and y axis labelling
plt.xlabel('Sports',fontsize=13)
plt.ylabel('Earning in US Dollars(million)',fontsize=13)
plt.xticks(rotation = 60) 
plt.show()#for displaying plot
```

* Plotting each sport earning using Seaborn

```python
sns.set_theme()
fig_dims = (15,8)
fig, ax = plt.subplots(figsize=fig_dims)
df_total_earning = df_ath.groupby('Sport')['Earning_mil'].sum().sort_values(ascending=False).reset_index()
plot = sns.barplot(x="Sport", y="Earning_mil", data=df_total_earning, color = 'darkorchid',ci=None, alpha=0.5)
ax.tick_params(axis='x', rotation=60)
# x and y axis labelling
plt.xlabel('Sports',fontsize=13)
plt.ylabel('Earning in US Dollars(million)',fontsize=13)
plt.title('Total produced earning for each sport',fontsize=13)
plt.show()#for displaying plot
```

From bar chart we can easily compare each sport earnings. Basketball, Boxing, Soccer and Golf have been the highest earnings sports comparative to the other sports in this dataset.

**4. What is the trend of overall earnings from the year 1990 to 2020?**

In this case, we are grouping data using 'Year' column by groupby() function and taking the average of earnings for each year using mean() function. Then plotting the data using matplotlib and seaborn 

* Plotting average athlete earning trend over the years using matplotlib

```python
ax = plt.axes()
ax.set_facecolor("gainsboro")
df_ath.groupby('Year')['Earning_mil'].mean().plot(kind = "line", color = 'royalblue',marker='o',fontsize=13,
                                        alpha=0.9,title = "Average athletes earning trend over years", figsize=(14,8))
# x and y axis labelling
plt.xlabel('Years',fontsize=13)
plt.ylabel('Earning in US Dollars(million)',fontsize=13)
plt.show() #for displaying plot
```

* Plotting average athlete earning trend over the years using Seaborn

```python
sns.set_theme()
fig_dims = (14,8)
fig, ax = plt.subplots(figsize=fig_dims)
x = df_ath['Year']
y = df_ath['Earning_mil']
p = sns.lineplot(x=x,y=y, data=df_ath, color = 'royalblue', ax=ax ,marker='o', ci=None)
# x and y axis labelling
plt.xlabel('Years',fontsize=13)
plt.ylabel('Earning in US Dollars(million)',fontsize=13)
plt.title('Average athletes earning trend over years',fontsize=13)
plt.show(); #for displaying plot
```

From the line chart we can easily observe the trend of earnings over the years & can clearly see that the overall earnings is seeing an upward trend

**5. Who all are the Top 10 paid athletes?**

For answering this research question first we are grouping data for 'Name' column using groupby() function and total earning using sum function. Then plotting horizontal bar chart using matplotlib and seaborn.

* Plotting total earning of top 10 athletes using matplotlib

```python
ax = plt.axes()
ax.set_facecolor("gainsboro")
df_name = df_ath.groupby('Name')['Earning_mil'].sum().sort_values(ascending = True).tail(10)
df_name.plot(kind = "barh",color = 'rebeccapurple',figsize=(14,8), width = 0.8, alpha = 0.5)
# x and y axis labelling
plt.xlabel('Name',fontsize=13)
plt.ylabel('Earning in US Dollars(million)',fontsize=13)
plt.title("Total produced earning for each sport",fontsize=13)
plt.show() #for displaying plot
```

* Plotting total earning of top 10 athletes using Seaborn

```python
sns.set_theme()
fig_dims = (14,8)
fig, ax = plt.subplots(figsize=fig_dims)
df_name = df_ath.groupby('Name')['Earning_mil'].sum().sort_values(ascending = False).reset_index().head(10)

p = sns.barplot(x= 'Earning_mil', y='Name', data=df_name, color = 'rebeccapurple', ax=ax ,orient = 'h', alpha=0.5, 
                ci=None,dodge=False)
# x and y axis labelling
plt.xlabel('Name',fontsize=13)
plt.ylabel('Earning in US Dollars(million)',fontsize=13)
plt.title("Total produced earning for each sport",fontsize=13)
plt.show() #for displaying plot
```

Sorting the bar chart with order makes value comparison easy. We can observe that 'Tiger Woods' has the highest earning in top 10 earning athletes.

**6. Who all are the Top 10 athletes appearing in the list maximum number of time.**

First we can taking the count of each name record using value_counts() method and taking the top 10 count values using head() function.

```python
df_name = df_ath['Name'].value_counts().sort_values(ascending=False).head(10)
df_name
```

* Plotting Top 10 athletes on the bases of appearance count in dataset using matplotlib

```python
ax = plt.axes()
ax.set_facecolor("gainsboro")
df_name = df_ath['Name'].value_counts().sort_values(ascending=True).tail(10)
df_name.plot(kind = "barh",color = 'midnightblue', figsize=(10,8), width = 0.8, alpha=0.5,fontsize=13)
# x and y axis labelling
plt.xlabel('No of apperence',fontsize=13)
plt.ylabel('Name of athlete',fontsize=13)
plt.title("Top 10 athletes names appearing in the list maximum times",fontsize=13)
plt.show() #for displaying plot
```

* Plotting Top 10 athletes on the bases of appearance count in dataset using seaborn

```python
sns.set_theme()
fig_dims = (10,8)
fig, ax = plt.subplots(figsize=fig_dims)
sns.countplot(y='Name', data=df_ath, order=df_ath.Name.value_counts().sort_values(ascending=False).head(10).index, 
              color = 'midnightblue',dodge=False,ax=ax,alpha=0.5)
# x and y axis labelling
plt.xlabel('No of apperence',fontsize=13)
plt.ylabel('Name of athlete',fontsize=13)
plt.title("Top 10 athletes names appearing in the list maximum times",fontsize=13)
plt.show() #for displaying plot
```

We can observe from bar plot that 'Michael Jordan' and 'Tiger Woods' have highest as well as equal number of appearances in the dataset.
